```python
#Loading the required Libraries
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
import seaborn as sns
```

```python
#Loading the ipl matches dataset
ipl=pd.read_csv('matches.csv')
```

```python
#having glace at first two record of the dataset
ipl.head()
```

```python
# looking at the number of rows and column
ipl.shape
```

```python
#getting the frequncy of most man of the match awards
ipl['player_of_match'].value_counts()
```

```python
# having glace at last five record of the dataset
ipl.tail()
```

```python
# appending the dataframe
rows=[pd.Series([636,637,2018,'mum','mi','csk','csk','bat','normal',0,'csk',35,0,'msd','chepauk','u1','u2','u3' ],index=ipl.columns),
     pd.Series([637,638,2018,'mum','mi','csk','csk','bat','normal',0,'csk',35,0,'msd','chepauk','u1','u2','u3' ],index=ipl.columns)]
ipl = ipl.append(rows, ignore_index=True)
```

```python

list(ipl['player_of_match'].value_counts()[0:5].keys())
```

```python
plt.figure(figsize=(8,5))
plt.bar(list(ipl['player_of_match'].value_counts()[0:5].keys()),(list(ipl['player_of_match'].value_counts()[0:5])),color='orange')
plt.show()                                                                 
```

```python
ipl['winner'].value_counts()[0:5].plot(kind='bar', color='green')
```

```python
#Visualize Count of season ipl matches
ipl['season'].value_counts().plot(kind='bar', color='red');
```

```python
ipl['season'].value_counts().plot(kind='barh', figsize=(8, 6), color='green')
plt.xlabel("season Frequency", labelpad=20)
plt.ylabel("Season", labelpad=14)
plt.title("Season Visualization", y=1);
```

```python

ipl['win_by_runs'].value_counts()
```

```python
#getting the record when a team won batting first
batting_first=ipl[ipl['win_by_runs']!=0]
```

```python
# lookin at the head
batting_first.head()
```

```python
# range by which teams are winning by batting first by histogram
plt.figure(figsize=(7,7))
plt.hist(batting_first['win_by_runs'],color='red')
plt.title("Distribution of runs")
plt.xlabel("runs")
plt.show()
```

```python
# finding out the number of wins w.r.t. each team after batting first
batting_first['winner'].value_counts()
```

```python
# making a bar plot top 3 teams with most wins after batting first
plt.figure(figsize=(8,5))
plt.bar(list(batting_first['winner'].value_counts()[0:3].keys()),(list(batting_first['winner'].value_counts()[0:3])),color=('blue','yellow','orange'))
plt.show()  
```

```python
plt.figure(figsize=(15,15))
plt.pie(list(batting_first['winner'].value_counts()),labels=(list(batting_first['winner'].value_counts().keys())),autopct='%0.1f%%')
plt.show()  
```
